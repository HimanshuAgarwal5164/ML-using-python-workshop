```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
#We will also see how to make pipelines, so also importing make_pipeline module.
from sklearn.pipeline import make_pipeline
```

```python
#We already know about scaling using MinMaxScaler module. Now we'll try scaling with StandardScaler module
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
```

```python
#We are using in-built datasets under sklearn module for practise 
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#After splitting our data into train and test datasets, we import our tree module as-
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
#Next we create the tree object
#It takes an optional 'max_depth' parameter which represents no. of layers of branching we'd like to see.

tree = DecisionTreeClassifier(max_depth=2)
```

```python
#After creating our tee object, we train it on training data
tree.fit(x_train,y_train)
```

```python
#Now that we've trained our Decision Tree, we'll plot it as-
plt.figure(figsize=(15,5))               #Describes size of image
plot_tree(tree, feature_names=cancer.feature_names, filled=True)              #Plots my tree
```

```python
#The most important thing to do is to decide the max_depth of a decision model. Because efficiency of my decision model is 
#heavily dependent on max_depth.
#One way of doing so is by manually training my model with different values for max_depth but it is impractical as 
#max_depth ranges from 1 to number of features.
#Another way of doing so is using GridSearchCV. Infact in real life applications, we have to only use GridSearchCV because 
#manually doing everything is impractical.
```

```python
#If had to do it manually-
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(x_train,y_train)
plt.figure(figsize=(15,5))
plot_tree(tree,feature_names=cancer.feature_names,filled=True)
#For this example, if we had to do it manually, we'd have to do this each time for max_depth=1 to max_depth=32(no. of 
#features).It is clearly dumb of us to even think about it
```

```python
for i in range (1,len(cancer.feature_names)+1):
    tree = DecisionTreeClassifier(max_depth=i)
    tree.fit(x_train,y_train)
    plt.figure(figsize=(15,5))
    plot_tree(tree,feature_names=cancer.feature_names,filled=True)
```

```python
#We've used a loop now to do my decision tree modelling for (max_depth=1 to max_depth=number of features).
#But this is time,memory consuming and impractical as stated by our Jupyter Notebook itself above.
#So,lets do it with GridSearchCv
```

## Doing decision tree modelling using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit

param_grid = {'max_depth':range(1,len(cancer.feature_names))}
grid_search = GridSearchCV(DecisionTreeClassifier(),param_grid,cv=StratifiedShuffleSplit(10))
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
pd.Series(grid_search.best_estimator_.feature_importances_ , index=cancer.feature_names).plot(kind='bar')
```

```python
plt.figure(figsize=(30,15))
plot_tree(grid_search.best_estimator_,feature_names=cancer.feature_names,filled=True)
```

```python
grid_search.score(x_test,y_test)
```

```python
y_pred = grid_search.predict(x_test)
```
